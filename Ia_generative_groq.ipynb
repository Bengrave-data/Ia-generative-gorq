```python
from google.colab import userdata
from groq import Groq

# Récupérer la clé API depuis les secrets (sécurisé)
api_key = userdata.get('GROQ_API_KEY')

# Créer le client Groq
client = Groq(api_key=api_key)

# Premier appel à l'IA
reponse = client.chat.completions.create(
    messages=[
        {
            "role": "user",
            "content": "Bonjour ! Présente-toi en 2 phrases."
        }
    ],
    model="llama-3.3-70b-versatile"
)

# Afficher la réponse
print(reponse.choices[0].message.content)
```

```python
def generer_post_linkedin(sujet):
    reponse = client.chat.completions.create(
        messages=[
            {
                "role": "system",
                "content": """Tu es un expert en marketing digital et copywriting.
Tu écris des posts LinkedIn percutants en français qui :
- Commencent par une accroche forte
- Font 3-4 paragraphes courts
- Se terminent par une question pour engager la communauté
- Utilisent des emojis avec parcimonie"""
            },
            {
                "role": "user",
                "content": f"Écris un post LinkedIn sur le sujet : {sujet}"
            }
        ],
        model="llama-3.3-70b-versatile",
        temperature=0.7
    )
    return reponse.choices[0].message.content

# Tester !
post = generer_post_linkedin("Comment j'ai appris l'IA en 2 semaines")
print(post)
```

```python
# Historique de conversation
historique = [
    {
        "role": "system",
        "content": "Tu es un assistant IA serviable et sympathique. Tu réponds en français de manière claire et concise."
    }
]

def chatter(message):
    # Ajouter le message de l'utilisateur à l'historique
    historique.append({"role": "user", "content": message})

    # Appeler l'IA
    reponse = client.chat.completions.create(
        messages=historique,
        model="llama-3.3-70b-versatile"
    )

    # Récupérer la réponse
    reponse_ia = reponse.choices[0].message.content

    # Ajouter la réponse à l'historique (pour garder le contexte)
    historique.append({"role": "assistant", "content": reponse_ia})

    return reponse_ia

# Tester la conversation
print("👤 Toi :", "Bonjour, je m'appelle Benjamin.")
print("🤖 IA :", chatter("Bonjour, je m'appelle Benjamin."))
print()
print("👤 Toi :", "Quel est mon prénom ?")
print("🤖 IA :", chatter("Quel est mon prénom ?"))
```
